## .apply() method on single columns

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\jamesbond.csv")
df.set_index(keys="Film",inplace=True)
df.sort_index(inplace=True)
```

```python
def mill(number):
    return str(number)+" Million"
    
df['Box Office']=df['Box Office'].apply(mill)
```

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\jamesbond.csv")
df.set_index(keys="Film",inplace=True)
df.sort_index(inplace=True)
```

## .apply() method on single row

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\jamesbond.csv")
df.set_index(keys="Film",inplace=True)
df.sort_index(inplace=True)
```

```python
df.head(10)
```

```python
def good_movie(row):
    actor=row[1]
    budget=row[4]
    if actor=="Sean Connery":
        return "Good Movie"
    elif actor=="Daniel Craig" and budget>100:
        return "okay movie"
    else:
        return "Waste movie"
df['Review']=df.apply(good_movie,axis="columns")
```
