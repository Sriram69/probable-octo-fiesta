## Expand and n paramters of .split() method

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\chicago.csv")
df['Department']=df['Department'].astype("category")
df['Employee Annual Salary']=df['Employee Annual Salary'].str.replace('$',"").astype("float")
df.dropna(how="all",inplace=True)
```

```python
df.head()
```

```python
df[["First Name","Last Name"]]=df['Name'].str.split(",",expand=True)
```

```python
df.head()
```

```python
df[["First Title Name","Reamining words"]]=df['Position Title'].str.split(" ",expand=True,n=1)
```
