## The .agg() method 

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\fortune1000.csv",index_col="Rank")
sector=df.groupby("Sector")
df.head()
```

```python
sector.mean().head()
```

```python
sector.agg({"Revenue":"sum",
             "Profits":"sum",
              "Employees":"mean"})
```

```python
sector.agg({"Revenue":["sum","mean"],
             "Profits":"sum",
              "Employees":"mean"}).head()
```

```python
sector.agg(["size","sum","mean"])
```

## Iterating through Groups 

### To get max Revenue/max Profits after GroupBy on any particular column

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\fortune1000.csv",index_col="Rank")
sector=df.groupby("Sector")
df.head()
```

```python
#But it does not return the entire row here
sector['Profits'].max()
```

```python
df_new=pd.DataFrame(columns=df.columns)
```

```python
df_new
```

```python
for sectors,data in sector:
    highest_rev_in_sector=data.nlargest(1,"Revenue")
    df_new=df_new.append(highest_rev_in_sector)
```

```python
df_new
```

```python
cities=df.groupby('Location')
```

```python
df_new_cities=pd.DataFrame(columns=df.columns)
```

```python
for city,data in cities:
    highest_rev_in_city=data.nlargest(1,"Profits")
    df_new_cities=df_new_cities.append(highest_rev_in_city)
```
