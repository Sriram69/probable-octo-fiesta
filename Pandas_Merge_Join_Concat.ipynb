## Merging by indexes, left_index and right_index parameter

```python
week1=pd.read_csv(r"E:\Data Analysis Pandas\Restaurant - Week 1 Sales.csv")
week2=pd.read_csv(r"E:\Data Analysis Pandas\Restaurant - Week 2 Sales.csv")
customers=pd.read_csv(r"E:\Data Analysis Pandas\Restaurant - Customers.csv",index_col=["ID"])
foods=pd.read_csv(r"E:\Data Analysis Pandas\Restaurant - Foods.csv",index_col=["Food ID"])
```

```python
customers.head()
```

```python
customers.loc[537]
```

```python
foods.head()
```

```python
sales=week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
sales.head(10)
```

```python
sales.merge(foods,how="left",left_on="Food ID",right_index=True)
```

```python
week1.merge(week2,how="left",right_index=True,left_index=True,suffixes=["-Week1","-Week2"])
```
