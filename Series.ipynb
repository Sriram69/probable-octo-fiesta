### The .get() method on a Series

```python
pokemon=pd.read_csv(r"E:\Data Analysis Pandas\pokemon.csv",index_col="Pokemon",squeeze=True)
pokemon.sort_index(inplace=True)
pokemon.head()
```

#### It is always a good idea to sort the index before doing any operations, so as to improve the speed

```python
pokemon.get([2,4])
```

#### If we give an index or a inddex label which is not present, then we will not get any error. It will just not display any output. This is actually due to the second parameter "default" inside the get method. The default value for "default " parameter is None

```python
# No error displayed
pokemon.get("ABCD")
```

```python
#If .get() does not find the key in the index, then it displays the default message
pokemon.get(key="ABCD",default="No pokemon found")
```

```python
pokemon.get(key="Aegislash",default="No pokemon found")
```

### The .apply() method

```python
google=pd.read_csv(r"E:\Data Analysis Pandas\google_stock_price.csv")
google.head()
```

```python
def performance(number):
    if number<300:
        return "Okay"
    elif number>=300 and number<=650:
        return "Average"
    else:
        return "Great"
```

```python
google['Stock Price'].apply(performance)
```

```python
google['Stock Price'].apply(lambda x:x+1)
```

```python
google.tail(5)
```
