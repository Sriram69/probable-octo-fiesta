### The .duplicated() method

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\employees.csv",parse_dates=["Start Date","Last Login Time"])
df['Senior Management']=df['Senior Management'].astype("bool")
df['Gender']=df['Gender'].astype("category")
df['Team']=df['Team'].astype("category")
df.head()
```

```python
df.sort_values(by="First Name",inplace=True)
```

```python
df.head(15)
```

#### The first row for each unique name will not be shown as duplicate, where as the other rows will be displayed, since they are
#### have duplciate names. The first row will be considered as original

```python
df.head(4)
```

```python
# Out of 1st 4 names for Aaron, the last 3 will be displayed as duplicate
df[df['First Name'].duplicated(keep="first")]
```

```python
# Keep=False marks all duplicates as True. SO in output we get all rows which are duplicates. Unique First Names will not
# be displayed
df[df['First Name'].duplicated(keep=False)]
```

```python
# Displaying unique records
df[~df['First Name'].duplicated(keep=False)]
```

### The .drop_duplicates() method

```python
import pandas as pd
df=pd.read_csv(r"E:\Data Analysis Pandas\employees.csv",parse_dates=["Start Date","Last Login Time"])
df['Senior Management']=df['Senior Management'].astype("bool")
df['Gender']=df['Gender'].astype("category")
df['Team']=df['Team'].astype("category")
df.head()
```

```python
df['Gender'].isnull().sum()
```

```python
df.sort_values(by="First Name",inplace=True)
```

```python
df.head(10)
```

```python
df.drop_duplicates(subset=['First Name'],keep="first").head(15)
```

```python
df.drop_duplicates(subset=['First Name'],keep=False)
```

```python
df.drop_duplicates(subset=["First Name","Team"],keep=False).head(10)
```
