### Rank values with the .rank() method

```python
nba=pd.read_csv(r"E:\Data Analysis Pandas\nba.csv").dropna(how="all")
nba.head()
```

```python
nba['Salary'].fillna(round(nba['Salary'].mean(),2),inplace=True)
nba.head(10)
```

```python
nba['Salary'].mean()
```

```python
nba['Salary_Rank']=nba['Salary'].rank(ascending=False).astype("int")
nba.head(10)
```

```python
nba.sort_values(by="Salary",ascending=False).head(10)
```

```python
nba.sort_values(by="Salary",ascending=True).head(10)
```
